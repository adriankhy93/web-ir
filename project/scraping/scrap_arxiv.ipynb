```python
import requests
from bs4 import BeautifulSoup
from PIL import Image
from io import BytesIO
import pandas as pd
```

```python
prof_info_df = pd.read_csv("data/prof_info.csv")
```

```python
def parase_url(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    return soup
```

```python
paper_urls_list = []
paper_titles_list = []
paper_abstracts_list = []

for name in prof_info_df["name"]:
    name_proc = "+".join(name.split())
    query_url = f"https://arxiv.org/search/?searchtype=author&query=%22{name_proc}%22&abstracts=show&size=100&order=-announced_date_first"
    print(name_proc)

    soup = parase_url(query_url)  # get html

    arxiv_link_block = soup.find_all(class_="list-title is-inline-block")
    paper_urls = [arxiv_link.a["href"] for arxiv_link in arxiv_link_block]
    paper_urls_list.append(paper_urls)

    arxiv_title_block = soup.find_all(class_="title is-5 mathjax")
    paper_titles = [arxiv_title.text.strip() for arxiv_title in arxiv_title_block]
    paper_titles_list.append(paper_titles)

    arxiv_abstract_block = soup.find_all(
        class_="abstract-full has-text-grey-dark mathjax"
    )
    arxiv_abstracts = [
        arxiv_abstract.text.strip()[:-10].strip()
        for arxiv_abstract in arxiv_abstract_block
    ]
    paper_abstracts_list.append(arxiv_abstracts)
```

```python
N = len(prof_info_df)
df_raw = []
for i in range(N):
    num_entries = len(paper_urls_list[i])
    for j in range(num_entries):
        entry = [
            i,
            paper_urls_list[i][j],
            paper_titles_list[i][j],
            paper_abstracts_list[i][j],
        ]
        df_raw.append(entry)
```

```python
new_paper_df = pd.DataFrame(
    df_raw, columns=["prof_id", "paper_link", "paper_title", "paper_abstract"]
)
```

```python
new_paper_df.to_csv("data/paper_data_NEW.csv", index=False)
```
